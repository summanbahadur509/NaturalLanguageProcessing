```python
"""
NLTK - Natural Language toolkit . This is a suit of libraries and programs for statistical natural language programing for 
English written in Python
"""
from nltk.tokenize import word_tokenize

def tokenizer_nltk(file_name):
    with open(file_name) as f:
        text = f.read()
        text = word_tokenize(text)
        "Tokenize with NLTK - commas and apostrophes are also considered as Tokens"
        print(text)
        
```

```python
"""
spaCy - When we need to tokenize text written in a language other the English , we can use spaCy. 
It supports more then 65 languages
"""

import spacy
def tokenizer_spaCy(file_name):
    with open(file_name) as f:
        text = f.read()
        nlp = spacy.load("en_core_web_sm")
        doc = nlp(text)
        "Spacy is object-oriented"
        for sentence in doc.sents:
            for word in sentence:
                print(word)
    
```

```python
tokenizer_spaCy(file_name)
```

```python
#Porter stemmer to Perform Stemming on each word
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
   
def porter_stemmer(file_name):
    ps = PorterStemmer()
    
    with open(file_name) as f:
        text = f.read()
        sent_text = nltk.sent_tokenize(text)
        # now loop over each sentence and tokenize it separately
        for sentence in sent_text:
            tokenized_sent = nltk.word_tokenize(sentence)
            for word in tokenized_sent:
                print(word, " : ", ps.stem(word))
```

```python
porter_stemmer(file_name)
```

```python
#Snowball stemmer
import nltk
from nltk.stem.snowball import SnowballStemmer
  
def snowball_stemmer(file_name):
    #the stemmer requires a language parameter
    snow_stemmer = SnowballStemmer(language='english')
    with open(file_name) as f:
        text = f.read()
        sent_text = nltk.sent_tokenize(text)
        for sentence in sent_text:
            tokenized_sent = nltk.word_tokenize(sentence)
            for w in tokenized_sent:
                print(w, " : ", snow_stemmer.stem(w))
```

```python
snowball_stemmer(file_name)
```

```python
#Write a program for spelling correction task using a look-up based approach.

import re
from collections import Counter

def words(text):
    return re.findall(r'\w+', text.lower())

# 1 - We need a LookUp table/dictionary of words from where we can check for the correct word
WORDS = Counter(words(open('big.txt').read()))

#2 - Using probabilities to select the most nearest correct word
def P(word, N=sum(WORDS.values())): 
    "Probability of `word`."
    return WORDS[word] / N

def correction(word): 
    "Most probable spelling correction for word."
    return max(candidates(word), key=P)

def candidates(word): 
    "Generate possible spelling corrections for word."
    return (known([word]) or known(edits1(word)) or known(edits2(word)) or [word])

def known(words): 
    "The subset of `words` that appear in the dictionary of WORDS."
    return set(w for w in words if w in WORDS)

#Getting all the possible words
def edits1(word):
    "All edits that are one edit away from `word`."
    letters    = 'abcdefghijklmnopqrstuvwxyz'
    splits     = [(word[:i], word[i:])    for i in range(len(word) + 1)]
    deletes    = [L + R[1:]               for L, R in splits if R]
    transposes = [L + R[1] + R[0] + R[2:] for L, R in splits if len(R)>1]
    replaces   = [L + c + R[1:]           for L, R in splits if R for c in letters]
    inserts    = [L + c + R               for L, R in splits for c in letters]
    return set(deletes + transposes + replaces + inserts)

def edits2(word): 
    "All edits that are two edits away from `word`."
    return (e2 for e1 in edits1(word) for e2 in edits1(e1))
```
